```python
# --------------------------------------------- 
# Synthetic Streaming Data Generator (Kafka Producer)
# ---------------------------------------------
# Purpose:
# Simulate real-time order events for SmartGear Retail system
# and push them to Kafka topic: 'smartgear_orders'
#
# Key Features:
# - Generates realistic order data
# - Covers last ~4 months using progressive timestamps
# - Enables Month-over-Month (MoM) analysis
# - Sends data in batches to simulate streaming ingestion
# ---------------------------------------------

from datetime import datetime, timedelta
import random
import json
import time
from confluent_kafka import Producer

# Kafka configuration (replace with your server details)
producer = Producer({
    'bootstrap.servers': 'pkc-921jm.us-east-2.aws.confluent.cloud:9092',
    'security.protocol': 'SASL_SSL',
    'sasl.mechanisms': 'PLAIN',
    'sasl.username': 'I73YUVFYOFS7YLR4',
    'sasl.password': 'cflti2Qkly5Sp2l3t0DzbmDugbhd26qndOLWUGai8WZ9MFavs1oaVorl9CRlJkqw'
})
# Sample reference data
products = ["Laptop", "Phone", "Tablet", "Headphones"]
regions = ["North", "South", "East", "West"]

# Realistic pricing logic
def get_price(product):
    if product == "Laptop":
        return round(random.uniform(40000, 120000), 2)
    elif product == "Phone":
        return round(random.uniform(10000, 80000), 2)
    elif product == "Tablet":
        return round(random.uniform(15000, 60000), 2)
    else:  # Headphones
        return round(random.uniform(1000, 10000), 2)

# Start from ~4 months ago
current_time = datetime.now() - timedelta(days=120)

# Infinite loop to simulate streaming
while True:
    batch_size = random.randint(5, 10)

    for _ in range(batch_size):
        product = random.choice(products)

        order = {
            "order_id": random.randint(1000, 9999),
            "timestamp": current_time.isoformat(),  # progressive timestamp
            "store_id": random.randint(1, 10),
            "product": product,
            "quantity": random.randint(1, 5),
            "price": get_price(product),
            "region": random.choice(regions)
        }

        producer.produce("smartgear_orders", json.dumps(order))
        print("Sent:", order)

        # Increment time gradually (important for MoM trends)
        current_time += timedelta(minutes=5)

    producer.flush()
    time.sleep(5)
```
